```python
import pandas as pd
import matplotlib.pyplot as plt # Import matplotlib

# Load the CSV file from Google Drive
file_path = '/content/drive/MyDrive/task1/API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip (Unzipped Files)/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, skiprows=4)  # skip metadata rows

# Display the first 5 rows of the dataset
print(df.head())

# Choose the year you're interested in
year = '2024'

# Filter relevant data (remove groups like "Africa Eastern and Southern")
filtered_df = df[df['Country Code'].apply(lambda x: len(x) == 3)]

# Select top 10 countries by population in that year
top_countries = filtered_df[['Country Name', year]].sort_values(by=year, ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_countries['Country Name'], top_countries[year])
plt.xlabel('Population')
plt.title(f'Top 10 Most Populous Countries in {year}')
plt.gca().invert_yaxis()  # Largest on top
plt.tight_layout()
plt.show()
```
